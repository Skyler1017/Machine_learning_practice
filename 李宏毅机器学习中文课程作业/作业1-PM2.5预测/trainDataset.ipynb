```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import math
```

```python
train = pd.read_csv("train.csv")
```

```python
train
```

```python
pm = train[train["testmaterial"]=="PM2.5"]
pm.drop(['date','station','testmaterial'], axis=1, inplace=True)
pm
```

```python
x = []
y = []
for i in range(15):
    tempx = pm.iloc[:,i:i+9]
    tempx.columns = np.array(range(9))
    tempy = pm.iloc[:,i+9]
    tempy.columns = np.array(range(1))
    x.append(tempx)
    y.append(tempy)
x = pd.concat(x)
y = pd.concat(y)
x = np.array(x, int)
y = np.array(y, int)
np.save("x.npy", x)
np.save("y.npy", y)
```
