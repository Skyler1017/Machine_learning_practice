```python
import os
import pandas as pd
import numpy as np
```

```python
train_path = r"/media/root/515e7d3a-49ac-40be-ba58-fef9702d123c/work_record/digit_recognizer/digit-recognizer/digit-recognizer/1216/digit/train"
val_path = r"/media/root/515e7d3a-49ac-40be-ba58-fef9702d123c/work_record/digit_recognizer/digit-recognizer/digit-recognizer/1216/digit/val"
label = np.loadtxt('label.csv')
```

```python
train_list = os.listdir(train_path)
train_label = []
train_img = []
num = len(train_list)
for i in range(num):
    prefix, type_ = os.path.splitext(train_list[i])
    if type_ == ".jpg":
        index = int(prefix)
        img_label = int(label[index-1])
        train_label.append(img_label)
        train_img.append(train_list[i])
img_s = pd.Series(train_img)
label_s = pd.Series(train_label)
df = pd.DataFrame()
df['img'] = img_s
df['label'] = label_s
path = os.path.join(train_path, 'dataset.csv')
df.to_csv(path, index=None, header=None)
```

```python
val_list = os.listdir(val_path)
val_label = []
val_img = []
num = len(val_list)
for i in range(num):
    prefix, type_ = os.path.splitext(val_list[i])
    if type_ == ".jpg":
        index = int(prefix)
        img_label = int(label[index-1])
        val_label.append(img_label)
        val_img.append(val_list[i])
img_s = pd.Series(val_img)
label_s = pd.Series(val_label)
df = pd.DataFrame()
df['img'] = img_s
df['label'] = label_s
path = os.path.join(val_path, 'dataset.csv')
df.to_csv(path, index=None, header=None)
```

```python
test_path = r"/media/root/515e7d3a-49ac-40be-ba58-fef9702d123c/work_record/digit_recognizer/digit-recognizer/digit-recognizer/1216/digit/test"
test_list = os.listdir(test_path)
len1 = len(test_list)
imgs_list = []
label_list = [i+1 for i in range(len1)]
for i in range(len1-1):
    file_name = str(i+1)+".jpg"
    imgs_list.append(file_name)
imgs_s = pd.Series(imgs_list)
label_s = pd.Series(label_list)
df = pd.DataFrame()
df['imgs'] = imgs_s
df['label'] = label_s
df.to_csv(test_path+'/dataset.csv', index=None, header=None)
```
