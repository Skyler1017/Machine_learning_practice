```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data.describe()
```

```python
data.loc[data['Sex']=='male', 'Sex']=1
data.loc[data['Sex']=='female', 'Sex']=0
```

```python
data['Age'].mean()
```

```python
data.loc[data['Age'].isnull(), 'Age']=29.69911
```

```python
data.isnull().sum()
```

```python
x = []
y = []
len1 = data.shape[0]
#sample = data.iloc[0, [2,4,5,6,7,9]]
#np.array(sample)
for i in range(len1):
    sample = data.iloc[i, [2,4,5,6,7,9]]
    x.append(np.array(sample))
    y.append(data.iat[i,1])
x = np.array(x)
y = np.array(y)
print (x)
print (y)
print (len(x))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.1, shuffle=False)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
import lightgbm as lgb
from xgboost import XGBClassifier
import time

rf = RandomForestClassifier()
ada = AdaBoostClassifier()
gbdt = GradientBoostingClassifier()
lgb = lgb.LGBMClassifier(num_leaves=127, n_estimators=200, max_depth=6, learning_rate=0.3, reg_alpha=0.05)
xgb = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.2)
model_list = ['RandomForest','AdaBoost','GBDT','XGboost','lightGBM']
for index, model in enumerate([rf, ada, gbdt, xgb, lgb]):
    a = time.time()
    model.fit(x_train, y_train)
    train_acc = model.score(x_train, y_train)
    test_acc = model.score(x_test, y_test)
    b = time.time()
    print ("{} train/test acc. is: {}/{}".format(model_list[index], train_acc, test_acc))
```
